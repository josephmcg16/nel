Format as an xml with the following sample schema:
```
<a:Channel>
  <a:CalculationModifier />
  <a:CanLogAsynchronously>true</a:CanLogAsynchronously>
  <a:Command>Modbus RTU Serial.4in USM.Block 1.SOS 2-3</a:Command>
  <a:Description>4in SOS 2-3</a:Description>
  <a:IsEnabled>true</a:IsEnabled>
  <a:MaximumLoggingFrequency>1</a:MaximumLoggingFrequency>
  <a:MultiplexerDelay />
  <a:ParentDevice z:Ref="i13" />
</a:Channel>
```

````python
import pandas as pd

df = pd.read_excel('USM Diagnostic Tags.xlsx', 'Sheet1')
my_xml = df.to_xml(
    root_name='Channels',
    row_name='Channel',
    index=False
)

# add namespace to column names
for col in df.columns:
    my_xml = my_xml.replace(
        col, f'a:{col}'
    )

# special cases
my_xml = my_xml.replace(
    'Channel', 'a:Channel')
my_xml = my_xml.replace(
    'True', 'true')
my_xml = my_xml.replace(
    'ParentDevice', 'ParentDevice z:Ref="i13" ')

with open('USM Diagnostic Tags.xml', 'w') as file:
    file.write(my_xml)
print(my_xml)
````
