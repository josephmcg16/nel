```python
import pandas as pd
import numpy as np
import serial
from datetime import datetime
import time
```

```python
def read_serial_line(com_port='COM5'):
    try:
        with serial.Serial('COM5', 9600) as ser:
            csv_string = ser.readline().decode().strip()
            data_line = np.asarray(csv_string.split(','), float)
    except (ValueError, serial.SerialException) as error:
        print(error)
        data_line = np.array([np.nan, np.nan])
    return data_line
```

```python
serial_data = []
try:
    while True:
        serial_data.append(
            [datetime.now().strftime(("%H:%M:%S.%f %d/%m/%Y"))] +
            list(read_serial_line())
        )
        print(*serial_data[-1])
except KeyboardInterrupt:
    pass
```

```python
df = pd.DataFrame(np.array(np.array(serial_data)), columns=['time', 'voltage (V)', 'current (mA)']).dropna()
df.shape
```

```python
import plotly.express as px

fig = px.line(df, x='time', y=['current (mA)', 'voltage (V)'])
fig.update_layout(template='plotly_dark', paper_bgcolor='#222222')
```
